### Mushroom
- 8124종류의 버섯의 특징과 독의 유무를 정리해논 데이터 셋
- 버섯 모양ㅇ의 특징을 기반으로 독의 유무를 판단
- 데이터 한줄이 버섯 한 종류
- 1번 컬럼 : 독의 유무(독(p), 식용(e))
- 2번 컬럼 : 버섯의 머리 모양    
b(벨형태), c(원뿔), x(볼록), f(평평), k(혹), s(오목)      
- 4번컬럼 :버섯의 머리 색    
n(갈색),b(황갈색), c(연한 갈색),g(회색)

#### V1컬럼 기준으로 p와 e를 정렬

```r
library(dplyr)
```

```r
mr <- arrange(mr, V1)
head(mr)
```

```r
tail(mr)
```

```r
# 정리한 dataset의 e와 p의 갯수 확인
# 식용버섯의 갯수
nrow(subset(mr, V1 == 'e'))
# 독버섯의 갯수
nrow(subset(mr, V1 == 'p'))
# 비율 
nrow(subset(mr, V1 == 'p')) / nrow(subset(mr, V1 == 'e'))
```

> V1 컬럼이 Target Column으로 합당

```r
library(caret)
```

```r
# 학습용과 검증용 분리하기
set.seed(1234)
samp <- createDataPartition(mr$V1, p=0.7, list=F)
data.train <- mr[samp,]
data.test <- mr[-samp,]
```

```r
head(data.test)
```

### RandomForest로 예측력 구하기

```r
library(randomForest)
```

```r
# 문제와 정답 만들기
x <- subset(data.test, select = -V1)
y <- data.test$V1
```

```r
# 학습시키기
rf <- randomForest(V1 ~ ., data = data.train, ntree = 100)
```

```r
# 예측하기
pred <- predict(rf, x)
```

```r
# 예측력과 오분류표
mean(y == pred)
table(y, pred)
```
